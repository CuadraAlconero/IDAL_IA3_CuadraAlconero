```
class Car:
  def __init__(self, engine, color):
    self.engine = engine
    self.color = color
    self._hidden_attribute = 'I am hidden =)'

  def _describe_car(self):
    print(f"El coche inicializado posee un motor {self.engine} y es {self.color}")

  def sell_car(self, assistant, price):
    print(f"Hola buenas tardes soy {assistant}")
    self._describe_car()
    print(f"Su precio es {price}")

  def random_method(self):
    pass
```

En ocasiones, es posible que interese construir un mtodo en una clase que solo dependa de informacin externa y no utilice ningn mtodo ni atributo de la clase. Estos mtodos son conocidos como mtodos estticos y se accede a ellos utilizando el decorator @staticmethod

```
class Car:
  def __init__(self, engine, color):
    self.engine = engine
    self.color = color
    self._hidden_attribute = 'I am hidden =)'

  def _describe_car(self):
    print(f"El coche inicializado posee un motor {self.engine} y es {self.color}")

  def sell_car(self, assistant, price):
    print(f"Hola buenas tardes soy {assistant}")
    self.describe_car()
    print(f"Su precio es {price}")

  @staticmethod
  def listen_music(song):
    print(f"La cancin: {song} est sonando en el coche")
```

En Python, toda variable que creis y funcin que defins es un objeto y por lo tanto podis interactuar con ellos como tal. Esto permite que definis interacciones de este tipo en este lenguaje:

```
class Calculator:
  def __init__(self, number_1, number_2):
    self.number_1 = number_1
    self.number_2 = number_2
    self.dict_functions = {"sum": self._sum, 
                           "substract": self._substract}

  def execute_calculations(self):
    iterable = self.dict_functions.items()
    for name_, fn in iterable:
      result = fn()
      print(type(fn))
      print(f"{name_}: {result}")

  def _sum(self):
    return self.number_1 + self.number_2

  def _substract(self):
    return self.number_1 - self.number_2
```

```
# Ejemplo de una factoria de zumos

class Apple:
  def __init__(self, weight):
    self.weight = weight

class Orange:
  def __init__(self, weight):
    self.weight = weight

class Blender:
  def __init__(self, list_apples, list_oranges):
    for apple in list_apples:
      if isinstance(apple, Apple):
        pass
      else:
        raise Exception("No todas las frutas son manzanas o naranjas")
    self.list_apples = list_apples
    for orange in list_oranges:
      if isinstance(orange, Orange):
        pass
      else:
        raise Exception("No todas las frutas son manzanas o naranjas")
    self.list_oranges = list_oranges
    self.orange_juice = 0
    self.apple_juice = 0

  def blend_fruit(self, type_fruit='apple'):
    if type_fruit == "apple":
      self._blend_apples()
    elif type_fruit == "orange":
      self._blend_oranges()
    else:
      raise Exception("Est licuadora solo extrae manzanas y naranjas")

  def _blend_apples(self):
    # De cada 2 kg de manzanas se extraen 1 de zumo
    qty = 0
    for apple in self.list_apples:
      qty += apple.weight
    self.list_apples = []
    self.apple_juice = qty/2

  def _blend_oranges(self):
    # De cada 3 kg de naranjas se extraen 1 de zumo
    qty = 0
    for orange in self.list_oranges:
      qty += orange.weight
    self.list_oranges = []
    self.orange_juice = qty/3

  def bottle_up(self, capacity_bottle, type_juice="apple"):
    if type_juice == "apple":
      bottles = self._bottle_apple_juice(capacity_bottle)
    elif type_juice == "orange":
      bottles = self._bottle_orange_juice(capacity_bottle)
    else:
      raise Exception("Esta licuadora solo contiene naranjas y manzanas")
    return bottles

  def _bottle_apple_juice(self, capacity):
    list_bottles = []
    while self.apple_juice >= capacity:
      bottle = Bottle(capacity, "apple")
      list_bottles.append(bottle)
      self.apple_juice += -capacity
    return list_bottles

  def _bottle_orange_juice(self, capacity):
    list_bottles = []
    while self.orange_juice >= capacity:
      bottle = Bottle(capacity, "orange")
      list_bottles.append(bottle)
      self.orange_juice += -capacity
    return list_bottles


class Bottle:
  def __init__(self, capacity, label):
    self.capacity = capacity
    self.label = label
```

```
list_apples = [Apple(10), Apple(5), Apple(4)]
list_oranges = [Orange(8), Orange(3), Orange(25)]

blender = Blender(list_apples, list_oranges)
```

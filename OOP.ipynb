El constructor de una clase es un método especial que se ejecuta al inicializar una instancia de cada clase. Su principal propósito es asignar los atributos de la clase y contener la información proporcionada en cada cápsula construída.

El concepto de instanciar una clase hace referencia a la construcción de un objeto utilizando el código desarrollado para la clase. Una buena forma de entender este proceso es utilizando una analogía entra un automóvil y el plano que define su construcción.

```
class Car:
  def __init__(self, engine, color):
    self.engine = engine
    self.color = color
```

El argumento self que aparece en el constructor de la clase hace referencia a la instancia que se construye (a si misma). De esta manera cuando nos referimos a los atributos self.engine = engine estamos asignando al objeto generado el valor que toma engine al inicializarse.

Un método de una clase permite implementar una función para modelar su comportamiento con su entorno. Definir un método en una clase en Python es relativamente sencillo, tan solo se necesita utilizar el comando def y añadir el argumento self al comienzo de la sentencia.

```
class Car:
  def __init__(self, engine, color):
    self.engine = engine
    self.color = color

  def describe_car(self):
    print(f"El coche inicializado posee un motor {self.engine} y es {self.color}")
```

En la práctica podéis diferenciar cuando accedáis a un atributo de un método comprobando si existen paréntesis al final de la sentencia. Un método de clase se utiliza como una función en Python y espera recibir los argumentos de entrada entre paréntesis. El argumento self se asigna automáticamente y no se debe definir cuando se utilice cualquiera de sus métodos, incluído el constructor de la clase.

```
car_1 = Car(engine="v8", color="azul")
car_1.describe_car()
```

Es posible controlar el uso y acceso que un usuario hace de una clase definiendo atributos y métodos ocultos. Este tipo de atributos y métodos no son accesibles a simple vista por un usuario y en principio utilizándolos se les indica que no deberían hacer uso de ellos.

Sin embargo, es común acceder a estos en la práctica y creo que se podrían considerar métodos y atributos a utilizar cuando se posea un conocimiento avanzado de la libreria o clase.

Para deinir un método o atributo oculto es necesario escribir un _ delante de su nombre

```
class Car:
  def __init__(self, engine, color):
    self.engine = engine
    self.color = color
    self._hidden_attribute = 'I am hidden =)'

  def _describe_car(self):
    print(f"El coche inicializado posee un motor {self.engine} y es {self.color}")

  def sell_car(self, assistant, price):
    print(f"Hola buenas tardes soy {assistant}")
    self.describe_car()
    print(f"Su precio es {price}")
```

```
car_1 = Car(engine="v8", color="azul")
```

Como se puede observar si ahora se escribe "car_1." los métodos y atributos ocultos no aparecen disponibles para su consulta. Sin embargo, es posible acceder a ellos si se escribe un "_".

En ocasiones, es posible que interese construir un método en una clase que solo dependa de información externa y no utilice ningún método ni atributo de la clase. Estos métodos son conocidos como métodos estáticos y se accede a ellos utilizando el decorator @staticmethod

```
class Car:
  def __init__(self, engine, color):
    self.engine = engine
    self.color = color
    self._hidden_attribute = 'I am hidden =)'

  def _describe_car(self):
    print(f"El coche inicializado posee un motor {self.engine} y es {self.color}")

  def sell_car(self, assistant, price):
    print(f"Hola buenas tardes soy {assistant}")
    self.describe_car()
    print(f"Su precio es {price}")

  @staticmethod
  def listen_music(song):
    print(f"La canción: {song} está sonando en el coche")
```

```
car = Car("v6", "red")
```

```
car.listen_music("Asereje")
```

En Python, toda variable que creáis y función que definís es un objeto y por lo tanto podéis interactuar con ellos como tal. Esto permite que defináis interacciones de este tipo en este lenguaje:

```
class Calculator:
  def __init__(self, number_1, number_2):
    self.number_1 = number_1
    self.number_2 = number_2
    self.dict_functions = {"sum": self._sum, "substract": self._substract}

  def execute_calculations(self):
    iterable = self.dict_functions.items()
    for name_, fn in iterable:
      result = fn()
      print(type(fn))
      print(f"{name_}: {result}")

  def _sum(self):
    return self.number_1 + self.number_2

  def _substract(self):
    return self.number_1 - self.number_2
```

```
calc = Calculator(6, 2)
```

```
# Ejemplo de una factoria de zumos

class Apple:
  def __init__(self, weight):
    self.weight = weight

class Orange:
  def __init__(self, weight):
    self.weight = weight

class Blender:
  def __init__(self, list_apples, list_oranges):
    for apple in list_apples:
      if isinstance(apple, Apple):
        pass
      else:
        raise Exception("No todas las frutas son manzanas o naranjas")
    self.list_apples = list_apples
    for orange in list_oranges:
      if isinstance(orange, Orange):
        pass
      else:
        raise Exception("No todas las frutas son manzanas o naranjas")
    self.list_oranges = list_oranges
    self.orange_juice = 0
    self.apple_juice = 0

  def blend_fruit(self, type_fruit='apple'):
    if type_fruit == "apple":
      self._blend_apples()
    elif type_fruit == "orange":
      self._blend_oranges()
    else:
      raise Exception("Está licuadora solo extrae manzanas y naranjas")

  def _blend_apples(self):
    # De cada 2 kg de manzanas se extraen 1 de zumo
    qty = 0
    for apple in self.list_apples:
      qty += apple.weight
    self.list_apples = []
    self.apple_juice = qty/2

  def _blend_oranges(self):
    # De cada 3 kg de naranjas se extraen 1 de zumo
    qty = 0
    for orange in self.list_oranges:
      qty += orange.weight
    self.list_oranges = []
    self.orange_juice = qty/3

  def bottle_up(self, capacity_bottle, type_juice="apple"):
    if type_juice == "apple":
      bottles = self._bottle_apple_juice(capacity_bottle)
    elif type_juice == "orange":
      bottles = self._bottle_orange_juice(capacity_bottle)
    else:
      raise Exception("Esta licuadora solo contiene naranjas y manzanas")
    return bottles

  def _bottle_apple_juice(self, capacity):
    list_bottles = []
    while self.apple_juice >= capacity:
      bottle = Bottle(capacity, "apple")
      list_bottles.append(bottle)
      self.apple_juice += -capacity
    return list_bottles

  def _bottle_orange_juice(self, capacity):
    list_bottles = []
    while self.orange_juice >= capacity:
      bottle = Bottle(capacity, "orange")
      list_bottles.append(bottle)
      self.orange_juice += -capacity
    return list_bottles


class Bottle:
  def __init__(self, capacity, label):
    self.capacity = capacity
    self.label = label
```

```
list_apples = [Apple(10), Apple(5), Apple(4)]
list_oranges = [Orange(8), Orange(3), Orange(25)]

blender = Blender(list_apples, list_oranges)
```
